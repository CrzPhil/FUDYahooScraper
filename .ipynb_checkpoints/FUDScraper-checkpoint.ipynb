```python
def getTrendingTickers():
    tickers = set()
    driver.get("https://finance.yahoo.com/trending-tickers")
    table = driver.find_element(By.TAG_NAME, "table")
    tbody = table.find_element(By.TAG_NAME, "tbody")
    rows = tbody.find_elements(By.TAG_NAME, "tr")
    for row in rows:
        tickers.add(row.text.strip().split('\n')[0])
    return tickers
```

```python
def getFinancials(ticker):
    out = []
    c_url = URL + f"quote/{ticker}/financials?p={ticker}"
    # Anti-blocking timeout. Tweak this but 5s works like a charm
    time.sleep(5)
    driver.get(c_url)
    try:
        header = driver.find_element(By.CLASS_NAME, "D\(tbhg\)")
        cols = header.find_elements(By.TAG_NAME, "span")
        table = driver.find_element(By.CLASS_NAME, "D\(tbrg\)")
        rows = table.find_elements(By.CSS_SELECTOR, "div[data-test='fin-row']")
    except Exception:
        return []
    out.append([col.text for col in cols])
    
    for row in rows:
        x = row.text.split('\n')
        x, y = x[0], x[1]
        tmp = [i for i in y.split(' ')]
        tmp.insert(0, x)
        
        out.append(tmp)
        
    return out
```

```python
def scrapeTickers(tickers: list, debug=False, resumeTicker=""):
    out = []
    errors = 0
    MAX_ERRORS = 10
    
    # Get ticker's index and continue scrape from there
    if resumeTicker != "":
        try:
            i = tickers.index(resumeTicker)
        except ValueError:
            return f"[Fatal] Ticker {resumeTicker} does not exist in your list."
    else:
        # Start from scratch
        i = 0
        
    for ticker in tickers[i:]:
        if errors >= MAX_ERRORS:
            print(f"{errors} reached. Aborting scrape; use a VPN/Proxychains and continue at {ticker}.")
            return out
        if debug:
            print(f"Current ticker: {ticker}")
        tmp = getFinancials(ticker)
        if tmp != []:
            out.append(tmp)
            if debug:
                print(f"Scrape for {ticker} successful.")
        else:
            errors += 1
            if debug:
                print(f"[Error] Scrape for {ticker} failed.\n Current Errors: {errors}/{MAX_ERRORS}")
    return out
```
